# Simple Model for the Beam-Profile from a Focused Transducer
A first estimate for th beam profile from a focused ultrasound transducer can be found from rather simple geometrical considerations by describeing the diffraction by the opening angle $\theta_0$ and the geometric focal length $F$. The model is taken from B. A. J. Angelsen "Ultrasound Imaging. Waves, Signals and Signal Processing", Emantec, 2000.

The transducer is described by its diameter $D$, geometric focal length $F$, and frequency $f$. The speed of sound in the medium is $c$. Parameters calculated from these are
\begin{align}
  &\text{Wavelength} & \lambda &= c/f \\
  &\text{Opening angle, two-sided} & \theta_0 &= 2 \lambda / D   & &\text{(-12 dB angle for a circular transducer) } \\
  &\text{F-number} & FN &= F / D     
\end{align}

A simple estimate for the beam shape can be found from these parameters by drawing four lines limiting the beam, 
- Two lines representing diffraction. These originate at the centre of the aperture and spread out by the opening angle $\theta_0$.
- Two lines representing the geometrical focus. These originate at the edges of the aperture, $x=\pm D$ and pass througt the geometric focus at $z=F$.

An estimate for the beam is then found from the outer limits of these four lines. This gives the following parameters fro the width, length and position of the focal zone
\begin{align}
   D_F &= \theta_0 F = 2 \lambda FN  & &\text{Beam diameter at focal length,  } z=F \\
   z_{F1} &= \frac{F}{1 + \theta_0 FN}  \quad , \qquad  z_{F2} = \frac{F}{1 - \theta_0 FN} & &\text {Focal zone limits} \\
   L_F &= z_{F2} - z_{F1} \approx 4 FN^2 \lambda = 2 FN D_F  & &\text {Focal zone length, the approximation is valid for $\theta_0  FN \ll 1$}
\end{align}
The width $D_F$ and length $L_F$ of the focal zone both increase with the f-number $FN$. This agrees with a low $FN$ meaning strong focusing.

Note that this is not ontended for accurate calculations of the beam profile, but as helpful esitmates when designing an ultrasound system. More accurate estimates can then be found from available simulation models such as  Field II.
