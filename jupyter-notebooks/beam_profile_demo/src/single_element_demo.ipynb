# Beam Profile From a Single Transducer Element 

## Radiated sound from a vibrating surface: The Rayleigh integral
The radiated pressure field $p(\vec r, t)$ from a vibrating surface $S_1$ in a rigid baffle is calculated from the Rayleigh integral, 
$$
p(\vec r,t)= \iint_{S_1} \frac {1}{2\pi R}  \frac {\partial v_n(t-R/c)}{\partial t }  dS_1 \:,
$$
where $s_1$ is the vibrating surface, described by coordinates $(x_1,y_1,z_1)$, $R$ is the distance from the source point $(x_1,y_1,z_1)$ to the field point $\vec r = (x,y,z)$, $v_n$ is component of the surface velocity normal to $S_1$, and $c$ is the speed of sound in the medium.

The calculations are more easily done in the time-domain, where the Rayleigh integral is transformed to
$$
\begin{aligned}
p(\vec r,t) &= Re\{ \hat p (\vec r,\omega) e^{j\omega t} \} &
\hat p(\vec r,\omega)&= \frac{j\rho c k v_0}{2\pi}\iint_{S_1} \frac {1}{R}  e^{j(\omega t-R/c)} \xi(x_1,y_1) dS_1 \: ,
\end{aligned}
$$
where $\rho$ is the density of the medium and $\rho c$ its characteristic acoustic impedance, $k=\omega /c$ is the wave number, and $v_0$ the normal component of vibrational velocity amplitude. $\xi(x_1,y_1)$ is called the apodisation function, it describes the relative variation of the vibration amplitude over the aperture $S_1$ so that $v_n(t)= Re\{ v_0 \xi(x_1,y_1) e^{j\omega t} \}$.

Note that short pulses, such as thosse used in medical ultrasound imaging, cover a broad frequency band and must be described by a range of frequencies. This smears out some of the phenomena found for a single  frequency, such as side lobes. However, the main properties concerning beam width and shape are the same as for single frequency pulses.

## Far-field approximation
Evaluation of the Rayleigh integral at any position in front of an arbitrary vibrating surface is complicated and can in general only be done numerically. However, approximate solutions can be found for some common aperture shapes in the far-field, i.e., at large distance from the source. The far-field approximations are also valid in the focal region of focused transducers. 

The Fraunhofer approximation is based on two assumptions

1. **Amplitude.** The variations in the amplitude factor $\frac{1}{2\pi R}$ are small when $R$ varies over the aperture $S_1=(x_1,y_1)$. These variations are ignored by setting $\frac{1}{2\pi R} \approx \frac{1}{2\pi r}$.

1. **Phase.** The phase factor $e^{-j\omega R/c}$  causes interference between waves from different positions on the aperture. These differences in $R$ need not be small compared to the wavelength $\lambda$ , and are included to the first order by the approximation. The distance $R$ from the source point $(x_1,y_1,z_1)$ to the field point $(x,y,z)$ can under this assumption be approximated to
$$
R= \sqrt{ (x-x_1)^2 -(y-y_1)^2 -z^2 )} \approx r - \frac{xx_1}{r} - \frac{yy_1}{r} = r-x_1 \sin\theta - y_1 \sin\phi \:.
$$
This expresses $R$ as a function of the distance $r$ from the aparture center and two direction angles $\theta$ and $\phi$.

Inserting this into the frequency domain version of the Rayleigh integral gives
$$
\begin{aligned}
\hat p(\vec r,\omega)& \approx \frac{j\rho c k v_0}{2\pi r} e^{-jkr}\iint_{S_1} e^{j(k_x x_1 + k_y y_1) } \xi(x_1,y_1) dS_1 \:.
\end{aligned}
$$

This result consists of three factors that can be interpreted as
1. $A(r)= \frac{j\rho c k v_0}{2\pi r}$ Amplitude decays proportional to distance, as $1/r$. This is the *diffaction loss*, the reduction in amplitude as the wave field spreads out.

1. $e^{-jkr}$                      Phase factor, corresponds to a spherical wave from the centre of the aperture.

2. $D(\theta,\phi) = \iint_{S_1} e^{j(k_x x_1 + k_y y_1) } \xi(x_1,y_1) dS_1$ Variation with angle, often called the *Dirctivity function*. $D(\theta,\phi)$ describes how narrow the beam is in tha azimuth ($\theta, x$) and elevation ($\phi, x$) directions.

Note that $D(\theta,\phi)$ is the 2D spatial Fourier-transformation of the apodization function $\xi(x_1,y_1)$. This allows fast and easy numerical calculationsby using the FFT-function built into most software tools. 

The integral for $D(\theta,\phi)$ can be solved analytically for simple apertures. Classic and important examples are rectangular and circular apertures with uniform oscillation amplitude, i.e. $\xi(x_1,y_1)=1$ inside the aperture $S_1$ and $\xi(x_1,y_1)=0$ outside $S_1$. See the lecture notes for details on how the calculations are evaluated.

1. Rectangular aperture of width $w$ and height $h$.
\begin{aligned}
D(\theta,\phi)& = \frac{1}{wh} sinc \left( \frac{w}{\lambda}\sin \theta \right) sinc \left( \frac{h}{\lambda}\sin \phi \right)  \: , &
sinc(u) &= \frac{\sin(\pi u)}{\pi u} \\
\end{aligned}

1. Cicrular aperture of diameter $d=2a$, where $a$ is the radius.
\begin{aligned}
D(\theta) & = \frac{2 J_1(k a \sin\theta)}{k a \sin\theta}
    = jinc \left( \frac{d}{\lambda} \sin\theta \right) 
      \: ,& jinc(u)&=  \frac{2J_1(\pi u)}{\pi u} \:,  & d&=2a \:,
\end{aligned}
where $J_1(x)$ is the Bessel function of the fist kind and order one. The jinc-function is introduced analogous to the sinc-function. Note that different definitions exist for the sinc- and jinc-functions, e.g., without the $\pi$ and the factor 2.

### Far-field limit, the *Rayleigh Distance*
The calculation above were done asuming the distance $R$ from a field pont $(x,y,z)$ to the aperture surface $(x_1,y_1,0)$ show suffieciently small variation over the aperture. An estimate of the minimum distance where this is valid is the *Rayleigh distance* $z_R$. This is the point of the acoustic axis where the distance $R$ to all points on the transducer aperture vary less than a quarter wavelength, i.e., the distance where $R_{max} - R_{min} = \tfrac14 \lambda$. The situation is illustrated in the figure below for a circular transducer with diameter $D$.
<div>
<img src="../fig/fig-rayleigh-distance.png" alt="Alternative text" width="800" />
</div>

The distance from $z_R$ to the centre of the aperture is simply $R_{min}=z_R$, while the distance from $z_R$ to the perimeter of the trasnducer is $R_{max}= \sqrt{z_R^2 + (\tfrac12 d)^2}$. 
Setting $z_R + \tfrac14 \lambda = R_{max}$ gives 
	\begin{align*}
		R_{max}^2 &= z_R^2 + \left(\tfrac{1}{2} d \right)^2 = \left(z_R +\tfrac{1}{4} \lambda \right)^2 
	\end{align*}
which gives the Rayleigh distance $z_R$ as
    \begin{align*}
    	z_R &= \frac{d^2}{2\lambda} + \tfrac{1}{8} \lambda \approx \frac{d^2}{2\lambda} \:.
	\end{align*}
$d \gg \lambda$ in most cases where this definitions are meaningful, so the last term can safely be neglected.
