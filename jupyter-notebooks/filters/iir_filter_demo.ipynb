# IIR Filter Demo

Calculate the resopnses of an IIR filter specified by its forward and backwards coeffients, $b$ and $a$.

- Time-domain: Impulse response
- Frequency domain: Magnitude and phase
- $z$-domains: Pole-zero plot


```python
#  Matplotlib backend to allow interactive operation
%matplotlib ipympl  
import ipywidgets
import filter_response    # Common filter response display and calculation
import iir_filter         # Specific for this demo

filt = filter_response.FilterResponse()

w = iir_filter.create_widgets()
display(w.layout)
res = ipywidgets.interactive(iir_filter.interact, 
                             filt = ipywidgets.fixed(filt),
                             a_text = w.widget['a'],
                             b_text = w.widget['b'])
```
