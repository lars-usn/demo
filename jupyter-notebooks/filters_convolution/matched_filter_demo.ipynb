# Matched Filter: Corralation of chirps (FM pulses)

The correlation between two signals $x[n]$ and $y[n]$  defined by
\begin{align}
  c[k] = \sum_{n=-\infty}^{n=\infty} x[n] y[n+k] 
\end{align}
The correlation $c[k]$ is a measure on how well the signal $y[n]$ matches $x[n]$ as function of the shift $k$.
