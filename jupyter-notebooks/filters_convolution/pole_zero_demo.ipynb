# Pole-zero Demo

Calculate the resopnses of an IIR filter specified by its poles and zeros.
The responses are illustrates as
- Time-domain: Impulse response
- Frequency domain: Magnitude and phase
- $z$-domains: Pole-zero plot

Poles and zeros are specified as complex numbers on polar form by their magnitude $r$ and angle $\theta$.

If the pole or zero is complex, i.e., the imaginary part is different from zero, the value is added to the diagram as a complex conjugate pair.
Real-valued poles and zeros are added as single values.


```python
#  Matplotlib backend to allow interactive operation
%matplotlib ipympl  
import ipywidgets
import filter_response    # Common filter response display and calculation
import pole_zero          # Specific for this demo

filt = filter_response.FilterResponse()

w, n_zp = pole_zero.create_widgets()
display(w.layout)
res = ipywidgets.interactive(pole_zero.interact, 
                             filt = ipywidgets.fixed(filt), 
                             n_zp = ipywidgets.fixed(n_zp), 
                             include_0=w.widget['include'][0], r_0=w.widget['r'][0], theta_0=w.widget['theta'][0],
                             include_1=w.widget['include'][1], r_1=w.widget['r'][1], theta_1=w.widget['theta'][1],
                             include_2=w.widget['include'][2], r_2=w.widget['r'][2], theta_2=w.widget['theta'][2],
                             include_3=w.widget['include'][3], r_3=w.widget['r'][3], theta_3=w.widget['theta'][3],
                             include_4=w.widget['include'][4], r_4=w.widget['r'][4], theta_4=w.widget['theta'][4],
                             include_5=w.widget['include'][5], r_5=w.widget['r'][5], theta_5=w.widget['theta'][5],
                             gain=w.widget['gain'])
```
