# Sampling Bandpass Signals

A signal $x(t)$ is sampled at time-points $t[n]= nT_s$, $n= 0, 1, 2, \ldots$, where $T_s$ is the sample interval and $f_s = 1/T_s$ is the sample rate. The resulting signal after sampling is $x[n] = x(n T_s)$.

The spectrum of $x[n]$ after sampling will be periodic, repeating itself at interval equal to the sampling rate $f_s$.

The phenomenon is illustrated in the interactive widget below. An example of a continous spectrum spectrum is shown in the upper graph. This spectrum is symmetric around $f=0$, as all spectra from a real signal, and extends to a bandwidth that can be set.

The middle graph showns the spectrum after sampling. The original spectrum (blue) will repeat itself with period $f_s$. The green area marks the interval inside the Nyquist limits, $\pm \tfrac{1}{2} f_s$. If the original spectrum extends outside the Nyquist limits, it will overlap with the reperated spectra centered at $f_s$ and $-f_s$, 

The lower graph shown the resulting spectra, which is the sum of the repeated spectra. If the original spectrum is inside the Nyquist limits, this is idential to the original spectrtum. However, if the spectrum extends outside the Nyquist limit, the overlap will cause the parts of the spectrum outsdide the Nyqist limit to fold down and appear inside the Nyquist limits, at an alias frequency.

The marker defined by the *Reference Frequency* shows frequencies that cannot be distinguished after sampling.
