## Rotational symmetry around 3-axis
This also implies no coupling betweeen shear and longitudinal strain 
$$
\begin{aligned}
 c_{11} &= c_{22} \neq c_{33} \\
 c_{12} &= c_{21} = c_{23} = c_{21} \neq c_{13} = c_{31}\\
 c_{44} &= c_{55} \neq  c_{66} \\
 c_{ij} &= 0 \text{  if  }  i=1,2,3 \text{ and } j= 4,5,6 
\end{aligned}
$$ 


$$
\begin{bmatrix}
T_1 \\
T_2 \\
T_3 \\
T_4 \\
T_5 \\
T_6 \\
\end{bmatrix}
=
\begin{bmatrix}
c_{11} & c_{12} & c_{13} & 0      &   0    & 0  \\
c_{12} & c_{11} & c_{13} & 0      &   0    & 0  \\
c_{13} & c_{13} & c_{33} & 0      &   0    & 0  \\
   0   &    0   &    0   & c_{44} &   0    & 0  \\
   0   &    0   &    0   &    0   & c_{44} & 0      \\
   0   &    0   &    0   &    0   &    0   & c_{66} \\
\end{bmatrix}
\begin{bmatrix}
S_1 \\
S_2 \\
S_3 \\
S_4 \\
S_5 \\
S_6 \\
\end{bmatrix}
$$

6 independent components, $c_{11}$, $c_{33}$, $c_{12}$, $c_{13}$, $c_{44}$, $c_{66}$.  
Written out 
$$
\begin{aligned}
 T_1 &= c_{11} S_1 + c_{12} S_2 + c_{13} S_3 \\
 T_2 &= c_{12} S_1 + c_{11} S_2 + c_{13} S_3 \\
 T_3 &= c_{13} S_1 + c_{13} S_2 + c_{33} S_3 \\

\end{aligned}
$$


## Isotropic material
$$
\begin{aligned}
 c_{11} &= c_{22} = c_{33} \\
 c_{12} &= c_{21} = c_{23} = c_{21} = c_{13} = c_{31}\\
 c_{44} &= c_{55} =  c_{66} \\
 c_{ij} &= 0 \text{  if  }  i=1,2,3 \text{ and } j= 4,5,6 
\end{aligned}
$$ 


$$
\left[\begin{array}{ccc}
T_1 \\
T_2 \\
T_3 \\
T_4 \\
T_5 \\
T_6 \\
\end{array}\right]
=
\begin{bmatrix}
c_{33} & c_{13} & c_{13} & 0      &   0    & 0  \\
c_{13} & c_{33} & c_{13} & 0      &   0    & 0  \\
c_{13} & c_{13} & c_{33} & 0      &   0    & 0  \\
   0   &    0   &    0   & c_{66} &   0    & 0  \\
   0   &    0   &    0   &    0   & c_{66} & 0      \\
   0   &    0   &    0   &    0   &    0   & c_{66} \\
\end{bmatrix}
\begin{bmatrix}
S_1 \\
S_2 \\
S_3 \\
S_4 \\
S_5 \\
S_6 \\
\end{bmatrix}
$$
3 independent components, $c_{33}$, $c_{13}$, and $c_{66}$.  
Written out 
$$
\begin{aligned}
 T_1 &= c_{33} S_1 + c_{13} S_2 + c_{13} S_3 \\
 T_2 &= c_{13} S_1 + c_{33} S_2 + c_{13} S_3 \\
 T_3 &= c_{13} S_1 + c_{13} S_2 + c_{33} S_3 \\
 T_4 &= c_{66} S_4  \\
 T_5 &= c_{66} S_5  \\
 T_6 &= c_{66} S_6  
\end{aligned}
$$
Thin rod

$$
\begin{aligned}
 T_1 &= T_ 2 = 0 \\
\end{aligned}
$$
$$
\begin{aligned}
 c_{33} S_1 + c_{13} S_2 + c_{13} S_3  &= 0  \\ 
 c_{13} S_1 + c_{33} S_2 + c_{13} S_3 &= 0   \\
 c_{13} S_1 + c_{13} S_2 + c_{33} S_3 &= T_3 \\
 \end{aligned}
$$
Define elastic constants for linear isotropic medium
$$
\begin{aligned}
  T_3 &= E S_3 & S_1 = S_2 &= -\frac{\nu}{E} S_3 \\
\end{aligned}
$$
Solving to find $S_3$ as function of $T_3$
$$
\begin{aligned}
 c_{33}(S_1+S_2) + c_{13}(S_1+S_2) + 2c_{13} S_3  & = 0 &   c_{13} (S_1+S_2) + c_{33} S_3 &= T_3 \\
 (c_{33} +  c_{13}) (S_1+S_2) + 2c_{13} S_3       &= 0  &   \\ \\
 (S_1+S_2) &= \frac{-2c_{13}}{c_{33} +  c_{13}}         &   \frac{-2c_{13}^2}{c_{33}+  c_{13}}  S_3+ c_{33} S_3 &= T_3 \\ \\
 \frac{c_{33}^2 - c_{13}^2}{c_{33}+  c_{13}}   S_3 = ( c_{33} - c_{13} )   S_3 &= T_3 \\ \\
 T_3  = (c_{33} - c_{13}) S_3 &= E S_3   &   c_{13} (S_1+S_2) + c_{33} S_3 &= T_3 \\
                                                 & &                    2c_{13} S_1  &= T_3 - c_{33} S_3 = \frac{1}{c_{33} - c_{13}}   S_3 + c_{33} S_3 = \frac{1-c_{33} + c_{13}}{c_{33} - c_{13}}   S_3 \\
 \end{aligned}
$$
