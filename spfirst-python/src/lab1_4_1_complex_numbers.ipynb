# 1)
Python code to enter $z_1=2 e^{j\pi/3}$ and  $z_2=-\sqrt{2} + 5j$

The numeric results give $z_1 = 1 + 1.73j$, $z_2 = -1.41 + 5j$, and $z_1+z_2 = -0.41 + 6.7j$.

This agrees with the phasor diagram.


```python
# Imports and defs
import numpy as np
import matplotlib.pyplot as plt
from cmath import pi, exp, sqrt

import zplot

pi_symbol = "\u03C0"

# efine a function to display complex numbers
def print_complex(z):
    """Print properties of the complex number z"""
    print(f"z = {z:.4f},     ", end="")
    print(f"Re{{z}} = {z.real:.3f},  Im(z) = {z.imag:.3f},  ", end="")
    print(f"|z| = {abs(z):.3f},  phase(z) = {np.angle(z)/pi:.3f} {pi_symbol}")

    return 0
    

#  Start program
z1 = 2*exp(1j*pi/3)
z2 = -sqrt(2) + 5j

print("z1")
print_complex(z1)
print()
print("z2")
print_complex(z2)
print()
print("z1+z2")
print_complex(z1+z2)

zplot.phasor([z1, z2, z1+z2], labels=[" $z_1$ ", " $z_2$ ", " $z_1+z_2$ "])
```

## 2)
The conjugate of a number $z=x+jy$ is $z^*=x-jy$, giving 

$Re\{z^*\} = Re\{z\} \quad Im\{z^*\}=-Im\{z\}, \quad |z^*|=|z|, \quad \angle z^*=-\angle z$

The inverse of a number $z= |z|e^{j\phi}$ is $z= |1/z|e^{-j\phi}$

$\quad |1/z|=1/|z|, \quad \angle 1/z=-\angle z$

The real and imaginary parts of $1/z$ are not so straightforward to identify.

These results seem to agree with the plots. Note that $z^*$ and $1/z$ have the same phase, $-\angle z$

```python
print("z1")
print_complex(z1)
print()
print("z1 conjugate")
print_complex(z1.conjugate())
print()
print("z1 inverse")
print_complex(1/z1)

zplot.phasor([z1, z1.conjugate(), 1/z1], labels=["$z_1$", "$z_1^*$", "$1/z_1$"])
plt.title("$z_1$, $z_1^*$, $1/z_1$")

print()
print("z2")
print_complex(z2)
print()
print("z2 conjugate")
print_complex(z2.conjugate())
print()
print("z2 inverse")
print_complex(1/z2)

zplot.phasor([z2, z2.conjugate(), 1/z2], labels=[" $z_2$ ", " $z_2^*$ ", " $1/z_2$ "])
plt.title("$z_2$, $z_2^*$, $1/z_2$")
```

## 3)
The magnitude of the product $z_1 z_2$ is product of the magnitudes of the two numbers, $|z_1 z_2|= |z_1 |z_2|$.

The phase is the sum of the phases of the two numbers, $\angle(z_1 z_2) = \angle(z_1)+\angle(z_2)$

The magnutude of the ratio $z_1/z_2$ is the ratio between the two magnitudes, $|z_1/z_2| = |z_1|/|z_2|$.

The phase is the difference between the phases of the two numbers, $\angle(z_1/z_2) = \angle(z_1)-\angle(z_2)$

This seems to agree with plots, although the difference vector is difficult to read.

```python
print("z1")
print_complex(z1)
print()
print("z2")
print_complex(z2)
print()
print("Product, z1*z2")
print_complex(z1*z2)
print()
print("Ratio, z1/z2")
print_complex(z1/z2)

zplot.phasor([z1, z2, z1*z2, z1/z2], labels=[" $z_1$ ", " $z_2$ ", " $z_1 z_2$ ", " $z_1 / z_2$ "])
```

## 4)
The product of conjugates is $z z^* = |z|^2$. 
This is a real number with magnitude equal to the square of the magnitude of $z$, as seen in the plots.

The ratio between conjugates is $z/z^* = e^{2j\phi}$. 
This is a complex number with magnitude 1 and angle $2\phi$.

The phasor plots agree with this.


```python
print("z1 ")
print_complex(z1)
print()
print("Product of conjugates, z1 z1*")
print_complex(z1*z1.conjugate())
print()
print("Ratio between , z1/z1*")
print_complex(z1/z1.conjugate())

zplot.phasor([z1, z1.conjugate(), z1*z1.conjugate(), z1/z1.conjugate()], labels=["$z_1$", "$z_1^*$", "$z_1 z_1^*$", "$z_1 / z_1^*$"])

print("z2 ")
print_complex(z2)
print()
print("Product of conjugates, z2 z2*")
print_complex(z2*z2.conjugate())
print()
print("Ratio between , z2/z2*")
print_complex(z1/z1.conjugate())

zplot.phasor([z2, z2.conjugate(), z2*z2.conjugate(), z2/z2.conjugate()], labels=["$z_2$", "$z_2^*$", "$z_2 z_2^*$", "$z_2 / z_2^*$"])
```

## 5)
The sum of two conjugates is $z+z^* = x + jy + x -jy = 2x= 2Re\{z\}$  
The difference between two conjugates is $z-z^* = (x + jy) - (x -jy) = 2jy= 2j Im\{z\}$.

This agrees with the phasor plots.


```python
print("5)")

print("Sum of conjugates, z1+z1*")
print_complex(z1*z1.conjugate())

zplot.phasor([z1, z1.conjugate(), z1+z1.conjugate(), z1-z1.conjugate()], labels=[" $z_1$ ", " $z_1^*$ ", " $z_1+z_1^*$ ", " $z_1 - z_1^*$ "])

print()
print("Sum of conjugates, z2+z2*")
print_complex(z2+z2.conjugate())

zplot.phasor([z2, z2.conjugate(), z2+z2.conjugate(), z2-z2.conjugate()], labels=[" $z_2$ ", " $z_2^*$ ", " $z_2+z_2^*$ ", " $z_2 - z_2^*$ "])
```
