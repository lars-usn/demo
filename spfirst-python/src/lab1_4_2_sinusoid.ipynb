The code to generate and plot a signal using the function 'make_cos' with the following parameters is shown below 

Amplitude $A=10^4$

Frequency $f_0=1.5 \times 10^6$Hz

Phase $\phi=-45^\circ = -\pi/4$

Duration $ 10^{-6}$s

The function is defined by 
$x(t)= A \cos(2 \pi f_0 t + \phi)$

For $t=0$, the value is 
$x(0)= A \cos(\phi) = A \cos(-\pi/4) = A \tfrac12\sqrt{2} = 0.71 \times 10^4 = 7100$

This value agrees with the plot.

The period is $T_0 = 1/f_0 = 0.667 \times 10^{-6}$ s. This agrees with the plot.

```python
import lab1

A = 1e4
f0 = 1.5e6
phi = np.radians(-45)
duration = 1e-6

x, t = lab1.make_cos(A, f0, phi, duration)

fig = plt.figure(figsize=([12, 4]))
ax = fig.add_subplot(1, 1, 1)
ax.plot(t,x)
ax.grid(True)
ax.set(xlabel = "Time [s]", 
       ylabel = "$x(t)$" )
```
