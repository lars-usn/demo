### 4.2 Python Function to Generate a Sinusoid Signal

# 1) 
Python code to generate a sinusoid with specified amplitude, frequency, phase and duration, sampled with 32 sam√•les per period

```python
import numpy as np
import matplotlib.pyplot as plt
from cmath import pi, exp     # For readability, also covered by numpy

def make_cos(A, f0, phase, duration):
    """Make a cosine-function from specified parameters.

    Parameters
    ----------
    A : float
        Amplitude
    f0: float
        Frequency [Hz]
    phase: float
        Phase [radians]
    duration: float
        Duration of signal [seconds}

    Returns
    -------
    x: 1D array of float
        Cosine-wave
    t: 1D array of float
        Time vactor [seconds]
    """

    T0 = 1/f0
    dt = T0/32
    t = np.arange(0, duration, dt)
    x = A * np.cos(2 * pi * f0 * t + phase)

    return x, t
```

## 2)
The code to generate and plot a signal using the function 'make_cos' with the following parameters is shown below 

Amplitude $A=10^4$

Frequency $f_0=1.5 \times 10^6$Hz

Phase $\phi=-45^\circ = -\pi/4$

Duration $ 10^{-6}$s

The function is defined by 
$x(t)= A \cos(2 \pi f_0 t + \phi)$

For $t=0$, the value is 
$x(0)= A \cos(\phi) = A \cos(-\pi/4) = A \tfrac12\sqrt{2} = 0.71 \times 10^4$

This value agrees with the plot.

```python
A = 1e-4
f0 = 1.5e6
phi = np.radians(-45)
duration = 1e-6

x, t = make_cos(A, f0, phi, duration)

plt.plot(t,x)
plt.grid(visible=True)
```
